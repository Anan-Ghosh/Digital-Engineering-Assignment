```python
from datetime import date
import requests
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Define the API base URL
BASE_URL = "https://api.carbonintensity.org.uk/intensity"

# Task 1: Fetch data from API
def fetch_from_to(start, end) -> list:
    url = f"{BASE_URL}/{start}/{end}"
    response = requests.get(url).json()
    return response["data"]

# Specify the date range
start_date = date(2023, 8, 1)
end_date = date(2023, 8, 7)

# Fetch data for the specified date range
data = fetch_from_to(start=start_date, end=end_date)

# Task 2: Create DataFrame and preprocess data
df = pd.DataFrame(data)
# Convert timestamps to datetime format
df['from'] = pd.to_datetime(df['from'])
df['to'] = pd.to_datetime(df['to'])
# A preview of the first 5 rows
display(df.head())
```

## **1.1 Descriptive Statistics** (8 pts)

Calculate and print the mean and standard deviation of carbon (actual) intensity values over the specified date range. **Round the output numbers to three decimal places**.


```python
# Your Contribution

# Extract actual intensity values
actual_intensity_values = [x['intensity']['actual'] for x in data]

# Calculate mean and standard deviation
mean_intensity = round(np.mean(actual_intensity_values), 3)
std_dev_intensity = round(np.std(actual_intensity_values), 3)

# Print the results rounded to three decimal places
print(f"Mean Intensity: {mean_intensity}")
print(f"Standard Deviation: {std_dev_intensity}")
```

**Explanation**

Using a list comprehension to extract the 'actual' intensity values from a list of dictionaries (data). It assumes that each dictionary in data has a key 'intensity', and under that key, there is a sub-key 'actual'. The extracted values are stored in the list actual_intensity_values.

np.mean(actual_intensity_values): Computes the mean (average) of the values in actual_intensity_values.
np.std(actual_intensity_values): Computes the standard deviation of the values in actual_intensity_values.
round(..., 3): Rounds both the mean and standard deviation values to three decimal places.

Then prints the calculated mean intensity and standard deviation. The f-string formatting is used to embed the calculated values into the printed strings. The results are rounded to three decimal places using the round function.

## **1.2 Time Period Analysis** (4 pts)

Calculate and print the duration of the data collection period (i.e., the time between the earliest and latest timestamps).

```python
# Your Contribution

#import datetime library
import datetime


# Extract timestamps
dates = [entry['from'][:10] for entry in data] + [entry['to'][:10] for entry in data]

# Convert dates to datetime objects
date_objects = [datetime.datetime.strptime(date, "%Y-%m-%d") for date in dates]

# Find the earliest and latest date
earliest_timestamp = min(date_objects)
latest_timestamp = max(date_objects)

# Calculate and print the duration
duration = latest_timestamp - earliest_timestamp
print(f"Data Collection Duration: {duration}")
```

## **1.5. Weekday vs. Weekend Analysis** (8 pts)

Create a new column indicating whether each entry corresponds to a weekday or weekend, and then calculate and print the mean (actual) carbon intensity values for weekdays and weekends separately. **Round the output numbers to three decimal places**.



```python
# Your Contribution

# Extract timestamps
timestamps = [x['from'] for x in data]

# Convert timestamps to datetime objects
datetime_objects = [datetime.datetime.fromisoformat(timestamp[:-1]) for timestamp in timestamps]

# Add a new column indicating whether each entry corresponds to a weekday or weekend
for entry, dt_object in zip(data, datetime_objects):
    day_of_week = dt_object.weekday()  # Monday is 0 and Sunday is 6
    entry['day_type'] = 'weekday' if day_of_week < 5 else 'weekend'

# Separate data into weekday and weekend entries
weekday_data = [entry for entry in data if entry['day_type'] == 'weekday']
weekend_data = [entry for entry in data if entry['day_type'] == 'weekend']

# Extract actual intensity values from weekday and weekend data
weekday_intensity_values = [entry['intensity']['actual'] for entry in weekday_data]
weekend_intensity_values = [entry['intensity']['actual'] for entry in weekend_data]

# Calculate mean intensity values for weekdays and weekends
mean_intensity_weekday = np.mean(weekday_intensity_values)
mean_intensity_weekend = np.mean(weekend_intensity_values)

# Print the results rounded to three decimal places
print(f"Mean Intensity (Weekday): {mean_intensity_weekday:.3f}")
print(f"Mean Intensity (Weekend): {mean_intensity_weekend:.3f}")
```

**Explanation**

Creates a list called timestamps using a list comprehension. It extracts the 'from' values (timestamps) from each entry in the data list.

Uses a list comprehension to convert each timestamp in the timestamps list to a datetime object using the fromisoformat method. "datetime.datetime:" This refers to the datetime class from the datetime module in Python's standard library. The [:-1] is used to remove the trailing 'Z' character from the ISO format timestamp.

Then iterates over each entry in data along with its corresponding datetime object. It calculates the day of the week using the weekday() method and adds a new key-value pair ('day_type') to each entry, indicating whether it corresponds to a weekday or a weekend.

Create two new lists (weekday_data and weekend_data) by filtering the entries based on the 'day_type' key added.

Use NumPy functions to calculate the mean intensity values for both weekdays and weekends.

Print the calculated mean intensity values for weekdays and weekends, rounded to three decimal places.

## **1.6 Time of Day Analysis** (8 pts)

Extract the hour of the day from timestamps and group data by the hour, and create a bar plot to visualize the mean carbon (actual) intensity for each hour of the day.

```python
# Your Contribution

# Extract timestamps and actual intensity values
timestamps = [x['from'] for x in data]
actual_intensity_values = [x['intensity']['actual'] for x in data]

# Convert timestamps to datetime objects and extract hours
datetime_objects = [datetime.datetime.fromisoformat(timestamp[:-1]) for timestamp in timestamps]
hours = [dt_object.hour for dt_object in datetime_objects]

# Create a dictionary to store mean intensity values for each hour
mean_intensity_by_hour = {}
for hour, intensity in zip(hours, actual_intensity_values):
    if hour not in mean_intensity_by_hour:
        mean_intensity_by_hour[hour] = [intensity]
    else:
        mean_intensity_by_hour[hour].append(intensity)

# Calculate mean intensity for each hour
mean_intensity_by_hour = {hour: np.mean(values) for hour, values in mean_intensity_by_hour.items()}

# Create a bar plot
hours_sorted = sorted(mean_intensity_by_hour.keys())
mean_intensities = [mean_intensity_by_hour[hour] for hour in hours_sorted]

plt.bar(hours_sorted, mean_intensities, color='skyblue')
plt.xlabel('Hour of the Day')
plt.ylabel('Mean Carbon Intensity')
plt.title('Mean Carbon Intensity for Each Hour of the Day')
plt.show()
```

```python
# Your Contribution

import cv2

# Load the image
image_path = "Kananaskis.jpg"
original_image = cv2.imread(image_path)


#Apply Gaussian blur
blurred_image = cv2.GaussianBlur(original_image, (7, 7), 0)

#Rotate the image counter-clockwise by 45 degrees
rows, cols, _ = original_image.shape
rotation_matrix = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)
rotated_image = cv2.warpAffine(original_image, rotation_matrix, (cols, rows))


#Resize the image to half of its original dimensions
resized_image = cv2.resize(original_image, (0, 0), fx=0.5, fy=0.5)

#Convert the image to grayscale
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

# Apply edge detection using edge detector
edges_image = cv2.Canny(gray_image, 200, 255)

#Binary thresholding to remove clouds
_, binary_threshold_image = cv2.threshold(gray_image, 80, 255, cv2.THRESH_BINARY)

# Display the images using Matplotlib
plt.figure(figsize=(12, 8))


# Gaussian blur
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('(a) Blurred Image')
plt.axis('off')

# Rotated image
plt.subplot(2, 3, 2)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('(b) Rotated Image')
plt.axis('off')

# Resized image
plt.subplot(2, 3, 3)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('(c) Resized Image')
plt.axis('off')

# Grayscale image
plt.subplot(2, 3, 4)
plt.imshow(gray_image, cmap='gray')
plt.title('(d) Grayscale Image')
plt.axis('off')


# Edge detection
plt.subplot(2, 3, 5)
plt.imshow(edges_image, cmap='gray')
plt.title('(e) Edges Image')
plt.axis('off')

# Binary thresholding to remove clouds
plt.subplot(2, 3, 6)
plt.imshow(binary_threshold_image, cmap='gray')
plt.title('(f) Binary Thresholding')
plt.axis('off')

plt.tight_layout()
plt.show()
```
