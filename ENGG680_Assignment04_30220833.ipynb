## **1.1 Histograms with KDE (Kernel Density Estimation)** (20 pts)

```python
# Your Contribution
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np


# Select the first 9 columns (variables)
variables= df.iloc[:, :9]


# Create a 3x3 grid of subplots
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
fig.suptitle('Histograms with KDE')

# List of variable names (replace with your variable names)
variables_names = ['Cement (component 1)(kg in a m^3 mixture)', 'Blast Furnace Slag (component 2)(kg in a m^3 mixture)', 'Fly Ash (component 3)(kg in a m^3 mixture)', 'Water  (component 4)(kg in a m^3 mixture)', 'Superplasticizer (component 5)(kg in a m^3 mixture)', 'Coarse Aggregate  (component 6)(kg in a m^3 mixture)', 'Fine Aggregate (component 7)(kg in a m^3 mixture)', 'Age (day)', 'Concrete compressive strength(MPa, megapascals)']


for i in range(9):
    row, col = i // 3, i % 3
    ax = axes[row, col]

    # Plot histogram
    hist = sns.histplot(variables.iloc[:, i], ax=ax, kde = True ,facecolor='lightgreen', color='red', bins=20)

    #adjust space between plots
    plt.subplots_adjust(hspace = 0.6, wspace = 0.6)

    #add the title
    ax.set_title(variables_names[i], fontsize=8)
    ax.set_xlabel('Value', fontsize=8)
    ax.set_ylabel('Density', fontsize=8)


# Add a title to the entire grid
plt.suptitle('Histograms with KDEs', fontsize=12)
plt.show()
```

## **1.3 Heatmap of Linear Correlations** (10 pts)

```python
# Your Contribution

# Calculate linear correlations between variables
correlation_matrix = variables.corr()


plt.figure(figsize=(10, 6))
# Create a heatmap with two decimal places and "RdYlGn" colormap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="RdYlGn", cbar=True, vmin=-1, vmax=1)
plt.title('Heatmap of Linear Correlations')


# Show the plots
plt.show()
```

## **2.1 Operating Revenue and Expenses (15 pts)**

```python
# Your Contribution
import seaborn as sns
import matplotlib.pyplot as plt


# Filter for "Operating revenue" or "Operating expenses" in "Summary statistics" and exclude "Canada" from the "GEO" column
revenue_expenses = df[(df["Summary statistics"].isin(["Operating revenue", "Operating expenses"])) & (df["GEO"] != "Canada")]

# Round all numbers to three decimal places
revenue_expenses = revenue_expenses.round({'VALUE': 3})

# Sort the DataFrame by "GEO"
revenue_expenses = revenue_expenses.sort_values(by="GEO")

display(revenue_expenses)

# Create a bar plot with Seaborn
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
ax = sns.barplot(data=revenue_expenses, x="GEO", y="VALUE", hue="Summary statistics", palette="muted")

#add axis title
plt.title("Operating Revenue and Expenses")
plt.xlabel("Province")
plt.ylabel("Value")
plt.legend(title="Summary statistics")

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Adjust y-axis limits for readability
max_value = revenue_expenses['VALUE'].max()
ax.set(ylim=(0, max_value + max_value * 0.1))



plt.show()
```

Selects data from the DataFrame df that meets specific criteria using isin and != conditions. Rounds the 'VALUE' column to 3 decimal places using round(3). Sorts the resulting data by 'GEO' column. Displays the processed data. plt.figure(figsize=(10, 6)): Initiates a new figure for plotting with a size of 10 inches in width and 6 inches in height. sns.barplot(): Utilizes Seaborn's barplot function to create a bar plot. data=revenue_expenses: Specifies the data to be used for plotting. x='GEO': Defines the 'GEO' column as the x-axis (province names). y='VALUE': Specifies the 'VALUE' column as the y-axis (revenue/expense values). hue='Summary statistics': Uses the 'Summary statistics' column for differentiating bars by revenue and expenses. palette="muted": Assigns a muted color palette for the bars. plt.title(): Sets the title of the plot as "Operating Revenue and Expenses". plt.xlabel(): Labels the x-axis as "Province". plt.ylabel(): Labels the y-axis as "Value". plt.legend(): Adds a legend with the title "Summary statistics" for the bar categories (revenue and expenses). ax.set_xticklabels(): Rotates the x-axis labels by 90 degrees for better readability. max_value = revenue_expenses['VALUE'].max(): Calculates the maximum value of 'VALUE'. ax.set(ylim=(0, max_value + max_value * 0.1)): Sets the y-axis limits to ensure all bars are clearly visible.

## **2.2 Salaries, Wages, Commissions, and Benefits (15 pts)**

```python
# Your Contribution

# Filter for "Salaries, wages, commissions and benefits" in "Summary statistics" and exclude "Canada" from the "GEO" column
salaries = df[(df['Summary statistics'] == 'Salaries, wages, commissions and benefits') & (df['GEO'] != 'Canada')]

# Drop rows with missing "VALUE" values
salaries = salaries.dropna(subset=['VALUE'])

# Round all numeric values to three decimal places
salaries = salaries.round({'VALUE': 3})

# Sort the DataFrame by "GEO"
salaries = salaries.sort_values(by='GEO')

display(salaries)

# Create a bar plot with Seaborn
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
ax = sns.barplot(data=salaries, x='GEO', y='VALUE', palette="muted")

# Add a title
plt.title("Salaries, Wages, Commissions, and Benefits")

# Add axis titles
plt.xlabel("Province")
plt.ylabel("Value")

# Rotate x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, horizontalalignment='right')

# Adjust y-axis limits for readability
max_value = salaries['VALUE'].max()
ax.set(ylim=(0, max_value + max_value * 0.1))

# Show the plot
plt.show()
```

data=salaries: Specifies the data to be used for plotting. x='GEO': Defines the 'GEO' column as the x-axis (provinces). y='VALUE': Specifies the 'VALUE' column as the y-axis (salaries, wages, commissions, and benefits). palette="muted": Assigns a muted color palette for the bars. Titles, Labels, and Customizations:

plt.title(), plt.xlabel(), and plt.ylabel(): Set the plot's title, x-axis label, and y-axis label, respectively. ax.set_xticklabels(): Rotates the x-axis labels for better readability. ax.set(ylim=(0, max_value + max_value * 0.1)): Adjusts the y-axis limits for a clear representation of data.
