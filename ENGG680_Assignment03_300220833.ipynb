**Tasks:**

**Report 1: Average Age of Patients**

**Report 2: Average Cholesterol Levels in Age Intervals**
- Divide patients into age intervals of 10 years each ([20, 30), [30, 40), [40, 50), [50, 60)).
- Calculate and report the average cholesterol (chol) levels for each age interval.

- Identify patients with the highest 30% and lowest 30% cholesterol levels in the dataset.

- Calculate and report the percentage of men and women with a positive diagnosis of heart disease (num=1).




```python
# =============================================================================
# Read the dataset
# =============================================================================
# The original dataset is available at the following link:
# https://www.kaggle.com/datasets/imnikhilanand/heart-attack-prediction
# For ease of access, the dataset has been uploaded on GitHub
url = 'https://raw.githubusercontent.com/HatefDastour/ENGG_680/main/Assignment_03/heart_data.csv'

import pandas as pd

# Load the data from 'heart_data.csv' into a DataFrame.
# Specify that there is no index column and the first row contains headers.
# Treat "?" as NaN (missing values).
df = pd.read_csv(url, na_values="?")

# Remove extra spaces before and after column names
df.columns = df.columns.str.strip()

# Display the first few rows of the DataFrame
display(df.head())
```

```python
# Your Contribution

#1
avg_age = df["age"].mean() #calculating the avarage age by mean from dataset
print("Average Age: ",avg_age) #print the avarage age
print('\n')

#2
bins = [20, 30, 40, 50, 60]
df['age_group'] = pd.cut(df["age"], bins,right = False) #define the age interval

ave_chol_age_group = df.groupby("age_group")["chol"].mean() #grouped by the age interval and calculate the avg chol in those age interval
print('Average Cholesterol by Age Group: ',ave_chol_age_group)  #print avarage Cholesterol by age group
print('\n')

#3
df = df.sort_values('chol') #sort the dataframe according to Cholesterol

highest = df.head((int(len(df)* (30/100))))
lowest = df.tail((int(len(df)* (30/100))))

#print the dataframe for highest 30% and lowest 30%
print("highest_30%\n")
display(highest)
print('\n\n')
print("lowest_30%\n")
display(lowest)


# Calculate and report the average resting blood pressure for these two groups
avg_trestbps_highest = highest['trestbps'].mean()
avg_trestbps_lowest = lowest['trestbps'].mean()

# print the results
print("Average trestbps for patients with the highest 30% cholesterol levels:", avg_trestbps_highest)
print("Average trestbps for patients with the lowest 30% cholesterol levels:", avg_trestbps_lowest)


#4
diagnoses_by_sex= df.groupby("sex")["num"].mean() #grouped by sex and calculate avg positive heart disease by sex

# Display the percentage of positive diagnoses for each gender
print("Percentage of men with a positive diagnosis of heart disease:" , diagnoses_by_sex[1] * 100, "%")
print("Percentage of women with a positive diagnosis of heart disease:", diagnoses_by_sex[0] * 100, "%")
```

$$
\text{CO2_categories}(x) =
\begin{cases}
\text{Low} & \text{if } 0 \leq x < 500 \\
\text{Medium} & \text{if } 500 \leq x < 1000 \\
\text{High} & \text{if } x \geq 1000

---
**Understanding Rolling Window:**
   A rolling window operation involves calculating a function (in this case, the cumulative sum) over a moving subset of data points within a specified window size. For example, with a window size of 3 hours, the cumulative sum will be calculated for every 3 consecutive hours of power consumption data. See this link for more details. https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rolling.html

---

```python
# Your Contribution

#1
for col in df.columns:     #ittrative through dataframe columns
  print(f"{col}_data_type = ", type(col))    #check the each columns type and print it


time_col_unique = len(df['time'].unique()) #Checking the length of unique values

print("Number of unique values in the time column ",time_col_unique) #print the length of unique values


#2
df['time'] = pd.to_datetime(df.time, format='%Y-%m-%d %H:%M:%S')   #converting time col into datetime format

df.head(10)

df = df.drop_duplicates() #drop duplicate rows

df.head()


#3

# Create a column to identify weekdays and weekends
df['Day_of_Week'] = df['time'].dt.dayofweek
df['Weekend'] = (df['Day_of_Week'] >= 5)

# Calculate the mean power consumption for weekdays and weekends separately
weekday_avg_power = df[~df['Weekend']]['power'].mean()
weekend_avg_power = df[df['Weekend']]['power'].mean()


# Create the 'CO2_category' column
df['CO2_category'] = pd.cut(df['CO2'], bins=[0, 500, 1000, float('inf')], labels=['Low', 'Medium', 'High'])


# Calculate the mean power consumption for each CO2 category
co2_avg_power = df.groupby('CO2_category')['power'].mean()
print(co2_avg_power)

#4

df.set_index('time', inplace=True)
df['Cumulative_Power_Consumption'] = df['power'].rolling('3H').sum()  # Calculate the cumulative sum of power consumption on a rolling 3-hour window


df['temp_change'] = df['temp'].diff()    # Calculate the change in temperature from the previous row

df.head()
```

   - For each of the following regions: **Canada** and **Alberta**
     - Filter the DataFrame to include only rows with the NAICS code "Engineering services [54133]" and the specified region.
     - Group the data by 'REF_DATE' and 'Summary statistics', and sum the 'VALUE' column.
     - [Pivot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pivot.html) the grouped data to create a summary table with 'Summary statistics' as the index and years as columns.

```python
# Your Contribution

#1
import pandas as pd

#Read the CSV file into a DataFrame
df = pd.read_csv("21100163.csv")


print("First 5 rows of the DataFrame:")
display(df.head())


#2

# Filter the DataFrame for Canada and Engineering services [54133]
canada_df = df[(df['GEO'] == 'Canada') & (df['North American Industry Classification System (NAICS)'] == 'Engineering services [54133]')]


# Group by 'REF_DATE' and 'Summary statistics', and sum the 'VALUE' column
canada_grouped = canada_df.groupby(['REF_DATE', 'Summary statistics'])['VALUE'].sum().reset_index()

# Pivot the grouped data
canada_pivot = canada_grouped.pivot(index='Summary statistics', columns='REF_DATE', values='VALUE')

# Move the row with 'Operating profit margin' to the last row
if 'Operating profit margin' in canada_pivot.index:
  operating_margin_row = canada_pivot.loc['Operating profit margin']
  canada_pivot = canada_pivot.drop(index='Operating profit margin')
  canada_pivot = canada_pivot.append(operating_margin_row)

# Display the summary table for Canada
print("\nSummary table for Canada:")
display(canada_pivot)



# Filter the DataFrame for Alberta and Engineering services [54133]
alberta_df = df[(df['GEO'] == 'Alberta') & (df['North American Industry Classification System (NAICS)'] == 'Engineering services [54133]')]

# Group by 'REF_DATE' and 'Summary statistics', and sum the 'VALUE' column
alberta_grouped = alberta_df.groupby(['REF_DATE', 'Summary statistics'])['VALUE'].sum().reset_index()

# Pivot the grouped data
alberta_pivot = alberta_grouped.pivot(index='Summary statistics', columns='REF_DATE', values='VALUE')


# Move the row with 'Operating profit margin' to the last row
if 'Operating profit margin' in alberta_pivot.index:
  operating_margin_row = alberta_pivot.loc['Operating profit margin']
  alberta_pivot = alberta_pivot.drop(index='Operating profit margin')
  alberta_pivot = alberta_pivot.append(operating_margin_row)


# Display the summary table for Alberta
print("\nSummary table for Alberta:")
display(alberta_pivot)
```

Data Analysis for Canada and Alberta

After filtering, the code groups the filtered data using groupby. It groups the data by 'REF_DATE' (time) and 'Summary statistics' columns.
