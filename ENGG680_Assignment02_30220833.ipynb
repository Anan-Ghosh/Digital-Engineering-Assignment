````python
class Account:
    def __init__(self, account_number, holder_name, balance):
        """
        Initializes an Account object.

        Parameters:
        - account_number (str): The unique account number.
        - holder_name (str): The name of the account holder.
        - balance (float): The initial account balance.

        This constructor initializes the account with the provided information.
        """
        # Your Contribution
        #initilize all the parameter into the constructor
        self.account_number = account_number
        self.holder_name = holder_name
        self.balance = balance

    def deposit(self, amount):
        """
        Deposits funds into the account.

        Parameters:
        - amount (float): The amount to be deposited.

        This method deposits the specified amount into the account balance,
        provided that the amount is greater than zero. It then prints the
        deposit transaction details.
        """
        # Your Contribution
        if (amount>0):   #checking the condition
          self.balance+= amount  #add the amount with initial balance
          print(f'Deposited ${amount}. New balance: ${self.balance}')

    def withdraw(self, amount):
        """
        Withdraws funds from the account.

        Parameters:
        - amount (float): The amount to be withdrawn.

        This method allows the account holder to withdraw the specified amount
        from the account balance, provided that the amount is greater than zero
        and does not exceed the current balance. It then prints the withdrawal
        transaction details.
        """
        # Your Contribution
        if (amount>0) and (amount<self.balance): #chceking two condition for withdrawing ammount from account
          self.balance = self.balance - amount    #substract withdraw amount from balance of the account
          print(f'Withdrew ${amount}. New balance: ${self.balance}')

    def display_account_info(self):
        """
        Displays account information.

        This function displays detailed information about the account,
        including account number, holder name, and current balance.
        """
        print("Account Information")
        print(f"Account Number: {self.account_number}")
        print(f"Account Holder: {self.holder_name}")
        print(f"Balance: ${self.balance}\n")

# =============================================================================
# Test the Account class using the following example
# =============================================================================
# Create accounts
account1 = Account("12345", "Alice", 1000.0)
account2 = Account("67890", "Bob", 500.0)

# Perform transactions
print('Perform transactions:')
account1.deposit(200.0)
account1.withdraw(150.0)

account2.deposit(100.0)
account2.withdraw(300.0)

# Display account information
print('\nDisplay account information:')
account1.display_account_info()
account2.display_account_info()
````

````python
class EnvironmentalDataAnalysis:
    def __init__(self, temp_data, precipitation_data):
        """
        Initialize an instance of the EnvironmentalDataAnalysis class with temperature and precipitation data.

        Parameters:
        - temp_data (list): List of daily temperature data in °C.
        - precipitation_data (list): List of daily precipitation data in mm.
        """
        #initilize all the parameter into the constructor
        self.temp_data = temp_data
        self.precipitation_data = precipitation_data

    def mean_temperature(self):
        """
        Calculate the mean temperature for the entire period.

        Returns:
        - float: The mean temperature in °C.
        """
        # Your Contribution
        length_temp = len(self.temp_data)  #check the length of temparature data and store into a varibale
        total_temp = sum(self.temp_data)   #sum all value of temparature data and store into a varibale named total_temp

        mean_temp = total_temp/length_temp  #get the mean temparature

        return mean_temp #return mean temparature

    def total_precipitation(self):
        """
        Calculate the total accumulated precipitation for the entire period.

        Returns:
        - float: The total precipitation in mm.
        """
        # Your Contribution
        length_particitation = sum(self.precipitation_data)  #sum all value of precipitation data and store into a varibale named length_particitation

        return length_particitation #return length_particitation

    def highest_temperature(self):
        """
        Find the highest daily temperature and its index.

        Returns:
        - tuple: A tuple containing the highest temperature in °C and its corresponding day index.
        """
        # Your Contribution
        max_temp = max(self.temp_data)        #measure max temparature from temparature data list
        max_temp_index = self.temp_data.index(max_temp)  #find the index number of max temparature

        return max_temp, max_temp_index

    def rainy_days(self, threshold):
        """
        Count the number of days with precipitation exceeding a threshold.

        Parameters:
        - threshold (float): The precipitation threshold in mm.

        Returns:
        - int: The number of rainy days exceeding the given threshold.
        """
        # Your Contribution
        rainy_days = 0  #define rainy_days variable with 0
        for days in self.precipitation_data:  #retrive data from precipitation_data
          if (days > threshold):  #chceking the condition
            rainy_days+=1         # count the number of rainy days in precipitation_data

        return rainy_days  #return rainy_days

    def analyze(self):
        """
        Perform overall environmental data analysis and print the results.
        """
        avg_temp = self.mean_temperature()
        total_precip = self.total_precipitation()
        max_temp, max_temp_index = self.highest_temperature()
        rainy_days_count = self.rainy_days(threshold=5.0)

        print("Environmental Data Analysis:")
        print(f"Mean Temperature: {avg_temp:.2f} °C")
        print(f"Total Accumulated Precipitation: {total_precip:.2f} mm")
        print(f"Highest Daily Temperature: {max_temp:.2f} °C (Day {max_temp_index + 1})")
        print(f"Number of Rainy Days (Precipitation > 5.0 mm): {rainy_days_count}")

# =============================================================================
# Example usage:
# The data is from Calgary Internatioanl Airport Weather Station
# =============================================================================

# Temperature data (°C) recorded for the period from September 1st to September 15th.
temperature_data = [24, 27.6, 21.4, 11.9, 22.2, 16.6, 19.2, 18.1, 21.2, 22.1, 24.5, 23.9, 20.9, 20.3, 25.7]

# Precipitation data (mm) collected for the period from September 1st to September 15th.
precipitation_data = [0, 0, 0, 6.6, 0.2, 1.7, 0, 0, 0, 0, 0, 0.4, 0, 0, 0]

# Create an instance of the EnvironmentalDataAnalysis class.
environmental_data = EnvironmentalDataAnalysis(temperature_data, precipitation_data)

# Perform data analysis using the EnvironmentalDataAnalysis instance.
environmental_data.analyze()
````

**Explanation**

mean_temperature(self):

This method calculates the mean (average) temperature from the temperature data stored in the self.temp_data list.It first calculates the length of the temperature data using len(self.temp_data).Then, it calculates the sum of all temperature values using sum(self.temp_data).Finally, it divides the total temperature sum by the length to obtain the mean temperature.The mean temperature is returned.


total_precipitation(self):

This method calculates the total accumulated precipitation from the precipitation data stored in the self.precipitation_data list.
It calculates the sum of all precipitation values using sum(self.precipitation_data).The total precipitation is returned.


highest_temperature(self):

This method finds the highest daily temperature and its corresponding day index from the temperature data.
It uses the max function to find the maximum temperature value from self.temp_data.Then, it finds the index of the maximum temperature using self.temp_data.index(max_temp).A tuple containing the highest temperature and its index is returned.


rainy_days(self, threshold):

This method counts the number of days with precipitation exceeding a specified threshold.It takes a threshold parameter as input, which represents the precipitation threshold in millimeters.It initializes a rainy_days counter variable to 0.It iterates through the precipitation data and checks if each day's precipitation exceeds the given threshold. If it does, it increments the rainy_days counter.The count of rainy days is returned.

````python
# Define a base class for students
class Student:
    def __init__(self, name, email, phone):
        """
        Initialize a Student object with name, email, and phone number.

        Parameters:
        - name (str): The name of the student.
        - email (str): The email address of the student.
        - phone (str): The phone number of the student.
        """
        # Your Contribution
        #initilize all the parameter into the constructor
        self.name = name
        self.email =  email
        self.phone = phone

    def display_info(self):
        """
        Display student information.
        """
        print(f"Name: {self.name}\nEmail: {self.email}\nPhone: {self.phone}")

    def update_email(self, new_email):
        """
        Update the email address of the student.

        Parameters:
        - new_email (str): The new email address to set.
        """
        # Your Contribution
        self.email = new_email #initialize email for updating

    def update_phone(self, new_phone):
        """
        Update the phone number of the student.

        Parameters:
        - new_phone (str): The new phone number to set.
        """
        # Your Contribution
        self.phone = new_phone  #initialize number for updating

# Define a subclass for university students inheriting from Student
class UniversityStudent(Student):
    def __init__(self, name, email, phone, student_id):
        """
        Initialize a UniversityStudent object with additional student ID.

        Parameters:
        - name (str): The name of the university student.
        - email (str): The email address of the university student.
        - phone (str): The phone number of the university student.
        - student_id (str): The student ID of the university student.
        """
        # Your Contribution
        super().__init__(name, email, phone)  #call super class constructor for initializing previous parameter
        self.student_id = student_id

    def display_info(self):
        """
        Display university student information including the student ID.
        """
        super().display_info()             #call super class function
        print(f"Student ID: {self.student_id}\n")

# Create a list to store university students
student_contact_book = []

# Function to add a university student to the contact book
def add_student(student_book, student):
    """
    Add a university student to the contact book.

    Parameters:
    - student_book (list): The list of university students.
    - student (UniversityStudent): The university student object to add.
    """
    student_book.append(student)

# Function to search for university students by name
def search_student_by_name(student_book, name):
    """
    Search for university students in the contact book by name.

    Parameters:
    - student_book (list): The list of university students.
    - name (str): The name to search for.

    Returns:
    - list: A list of matching university student objects.
    """
    matching_students = []
    for student in student_book:
      # Your Contribution
      if name in student.name:    #matching the name from the student
        matching_students.append(student)   #adding into the list
    return matching_students         #return the list of matching students

# Instantiate UniversityStudent objects
alice = UniversityStudent("Alice Johnson", "alice.johnson@university.com", "123-456-7890", "S12345")
bob = UniversityStudent("Bob Smith", "bob@university.com", "555-123-4567", "S67890")
charlie = UniversityStudent("Charlie Brown", "charlie@university.com", "555-555-5555", "S45678")

# Add university students to the contact book
add_student(student_contact_book, alice)
add_student(student_contact_book, bob)
add_student(student_contact_book, charlie)

# Update student information
alice.update_email("alice.johnson_new_email@university.com")
charlie.update_phone("555-999-9999")

print(f"All Contacts:")
# Display contact information
for contact in student_contact_book:
    contact.display_info()

# Search for students by name
search_name = "Alice"
matching_students = search_student_by_name(student_contact_book, search_name)

# Display information of matching students
print(f"Matching contacts for '{search_name}':")
for student in matching_students:
  student.display_info()
````
